### Funzioni per la manipolazione dei dati

#### Funzione che prende in input prima train poi test ed effettua il join della tabella stats con le caratteristiche dei pokemon in battaglia (si nota come le caratteristiche del secondo abbiano il "second" come suffisso)

```python
def join_with_stats(dataFrame):
    dataFrame = dataFrame.join(stats.set_index("#"), on="First_pokemon", rsuffix="_first")
    dataFrame = dataFrame.join(stats.set_index("#"), on = "Second_pokemon", rsuffix="_second")
    return dataFrame
```

#### Funzione che effettua il replace del winner. Se ha vinto il primo allora "0" altrimenti "1"

```python
def replace_winner(dataFrame):
    dataFrame.Winner[dataFrame.Winner == dataFrame.First_pokemon] = 0
    dataFrame.Winner[dataFrame.Winner == dataFrame.Second_pokemon] = 1
    return dataFrame
```

#### Funzione per l'analisi dei dati. Viene creata una Heat-map per osservare la correlazione delle feature

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
def heat_map():
    correlation = train.corr()
    plt.figure(figsize=(30,30))
    sns.heatmap(correlation, square=True,annot=True,cmap='cubehelix')
    plt.title('Correlation between different features')
```

```python
train = replace_winner(train)
train = join_with_stats(train)
```

```python
heat_map()
```

#### Si decide di aggiungere delle colonne che rappresentano la differenza in termini di "punteggio delle abilità" tra il primo e il secondo pokemon 

```python
def cal_diff(dataFrame):
    dataFrame['HP_Diff'] = dataFrame['HP'] - dataFrame['HP_second']
    dataFrame['Atk_Diff'] = dataFrame['Attack'] - dataFrame['Attack_second']
    dataFrame['Def_Diff'] = dataFrame['Defense'] - dataFrame['Defense_second']
    dataFrame['Sp_Atk_Diff'] = dataFrame['Sp. Atk'] - dataFrame['Sp. Atk_second']
    dataFrame['Sp_Def_Diff'] = dataFrame['Sp. Def'] - dataFrame['Sp. Def_second']
    dataFrame['Speed_Diff'] = dataFrame['Speed'] - dataFrame['Speed_second']
    return dataFrame
```

#### Dopo l'analisi della mappa sopra visualizzata, si ritiene utili mantenere solamente le colonne appena inserite, e si droppano le colonne ritenute poco utili

```python
def drop_columns(dataFrame):
    columns_to_drop = ['Attack', 'Attack_second', 'Defense', 'Defense_second', 'HP', 'HP_second',
                   'Sp. Atk', 'Sp. Atk_second', 'Sp. Def', 'Sp. Def_second', 'Speed', 'Speed_second',
                   'Legendary', 'Legendary_second', 'First_pokemon', 'Second_pokemon','Generation',
                       'Generation_second', 'Type 1','Type 2', 'Type 1_second','Type 2_second','Name','Name_second']
    dataFrame = dataFrame.drop(columns=columns_to_drop)
    return dataFrame
```

```python
train = cal_diff(train)
train = drop_columns(train)

test = replace_winner(test)
test = join_with_stats(test)
test = cal_diff(test)
test = drop_columns(test)
```

```python
y_train = train["Winner"].copy()
X_train = train.drop("Winner",axis=1).copy()

y_test = test["Winner"].copy()
X_test = test.drop("Winner",axis=1).copy()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
trainedtree = model.fit(X_train, y_train)

pred = model.predict(X_test)

print(accuracy_score(y_test, pred)*100)
print(model.score(X_test, y_test))
```
